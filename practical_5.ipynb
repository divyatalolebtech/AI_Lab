```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import emoji
```

```python
# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Load your dataset
df = pd.read_csv("instagram_comments_curated_dataset.csv")
```

```python
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def preprocess_comment(text):
    text = re.sub(r'@\w+|#\w+|http\S+', '', str(text))
    text = emoji.demojize(text, language='en')
    tokens = text.split()
    tokens = [stemmer.stem(w.lower()) for w in tokens if w.isalpha() and w.lower() not in stop_words]
    return ' '.join(tokens)
```

```python
df["ProcessedComment"] = df["Comment"].astype(str).apply(preprocess_comment)
df.to_csv("practical5_preprocessed_output.csv", index=False)
df[['Comment', 'ProcessedComment']].head()
```
